```python
import requests
import io
import pandas as pd
import numpy as np
```

# 1. import phenology dataset and select data

```python
#import phenology (climate+daylength+observations) data from GitHub:Floydworks
url = ('https://raw.githubusercontent.com/Floydworks/WildflowerFinder_Phenology_Tool/main/cleaned_data_files/phenology_dataset_2017_2022_df.csv')
download = requests.get(url).content
# Read the downloaded content and turn it into a pandas dataframe
phenology_data = pd.read_csv(io.StringIO(download.decode('utf-8')))
print('phenology_data:', phenology_data.shape)

```

```python
df_test = phenology_data[phenology_data['park']=='MtDiablo']
print(df_test.shape)
df_test
```

# 2. manually label observations for Adelinia grande

### 2a. export unlabeled data and label it

```python
#reduce columns and export the test data for labeling
df_test_url_export = df_test[['id', 'park', 'plain_dates', 'genus_species', 'url', 'image_url']]

#export csv for manually checking observation URLs for bloom or not in bloom
df_test_url_export.to_csv('/Users/sandidge/Desktop/df_test_url_export.csv')
```

### 2b. select data for species of interest

```python
df = df_test

#get observation counts for each species
spp_count = df.groupby(['genus_species'])['genus_species'].count().sort_values(ascending=False).reset_index(name='count')
print(spp_count.head(16))
```

```python
df = df[df['genus_species']== 'Adelinia grande']
df.head(3)
```

### 2c. import labeled data (will be updated with more species labeled)
df_test_url_export ---> mtdiablo_test

```python
#import phenology (climate+daylength+observations) data from GitHub:Floydworks
url = ('https://raw.githubusercontent.com/Floydworks/WildflowerFinder_Phenology_Tool/main/cleaned_data_files/test_mtdiablo_partially_labeled.csv')
download = requests.get(url).content
# Read the downloaded content and turn it into a pandas dataframe
test_data_labeled = pd.read_csv(io.StringIO(download.decode('utf-8')))
print('test_data_labeled:', test_data_labeled.shape)

#get only labeled observations
df_test_labeled = test_data_labeled[~test_data_labeled['status'].isnull()]
print(df_test_labeled.shape)
df_test_labeled.head()

df_test_labeled['genus_species'].unique()
```

### 2d. check labels and map labeled status onto the main dataframe

```python
#check for additional unlabeled observations of species of interest
#list of observation id that are labeled
df_test_labeled_ids = list(df_test_labeled['id'].unique())
print('labeled data: ',len(df_test_labeled_ids))

#list of all ids in the current dataframe
df_ids = list(df['id'].unique())
print('all data: ',len(df_ids))

#just the rows of unlabeled observations
df_unlabeled = df[~df['id'].isin(df_test_labeled_ids)]

print('unlabeled data: ',len(df_unlabeled))
if len(df_test_labeled) == len(df_ids):
    print('\nNo new data to label!')
```

```python
#create dictionary of main observation ids and status, old dataset
id_status_dict_main = dict(zip(df_test_labeled['id'], df_test_labeled['status']))
print(len(id_status_dict_main))

#map new labels onto phenology dataset
df['status'] = df['id'].map(id_status_dict_main)
#print(df.shape)

#check for remaining missing labels
#print('There are: ',df['status'].isna().sum(), '  missing status labels in the dataset')
#df.head()
```
